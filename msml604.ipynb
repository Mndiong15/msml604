```python
# Essential packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Create a 100x100 "image"
height, width = 100, 100
variance_map = np.full((height, width), 0.1)  # Low variance everywhere
# 2. Introduce high-variance regions
# Middle noisy patch
variance_map[30:70, 30:70] = 1.0

# Noisy patch in the top-right corner
variance_map[10:20, 80:90] = 2.0
```

```python
# 3. Visualize the variance map
plt.figure(figsize=(6,6))
plt.imshow(variance_map, cmap='hot', interpolation='nearest')
plt.colorbar(label='Initial Variance')
plt.title('Simulated Variance Map')
plt.xlabel('X-axis (pixels)')
plt.ylabel('Y-axis (pixels)')
plt.show()
```

```python
# 4. Set total sample budget
total_samples = 10000  # Fixed total number of samples

print(f"Total sample budget: {total_samples}")
print(f"Variance map shape: {variance_map.shape}")
```
